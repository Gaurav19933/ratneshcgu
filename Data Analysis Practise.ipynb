```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N=[*range(0,1000,20)]
T=[*range(0,100,2)]
print(N,len(N))
print(T,len(T))
```

```python
power=[]
for i in range(len(T)):
    power.append(2*3.14*N[i]*T[i]/((60)*1000))
print(power,"kW")
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x=power,y=N)
sns.lineplot(x=power,y=T)
plt.xlabel("KW")
plt.ylabel("RPM")
plt.ylabel("N.m")
plt.legend('ABCDEF', ncol=2, loc='upper left')
plt.legend( ("N","T"),ncol=2, loc='upper left')
plt.show()
```
