```python
import numpy as np
import matplotlib.pyplot as plt
phase_1=[]
phase_2=[]
phase_3=[]
vmax=180*(np.pi/180)
radians=[]
degree=[]
vsum=[]
for i in range (0,390):
    degree.append(i)
    radians.append(i*(np.pi/180))
for j in range(390):
    vinst_1=vmax*np.sin(radians[j])
    phase_1.append(vinst_1)
    vinst_2=vmax*np.sin(radians[j]-(120*(np.pi/180)))
    phase_2.append(vinst_2) 
    vinst_3=vmax*np.sin(radians[j]-(240*(np.pi/180)))
    phase_3.append(vinst_3)  
for k in range (390):
    vcoll=phase_1[k]+phase_2[k]+phase_3[k]
    vsum.append(vcoll)
#power calculation :
#print(radians)
#print(len(radians))
#print(degree)
#print(len(degree))
#print(vmax)
#print(phase_1)
print(vsum,"Vsum length :",len(vsum),"degree length : ", len(degree),"minimum value :",min(vsum))
plt.plot(degree,phase_1)
plt.plot(degree,phase_2)
plt.plot(degree,phase_3)
plt.plot(degree,vsum)
plt.show()
```

```python
# convert feet and inches to cm

def cm(feet=0,inches=0): # with key arguments elements can be exchanged
    feet_cm=feet*12*2.5
    inches_cm=inches*2.5
    return (feet_cm+inches_cm)

# cm(input("feet:",feet),input("inches:",inches))

print(cm(feet=5))
print(cm(inches=7))
print(cm(feet=5,inches=7))
```

```python
# sort and add to get summation curve
formlist=[]
a=[4,2,1,3]
b=[0.5,1,2.5,5,10]
c=[1,2,8,7,6,3,8,9,10,15]
d=[0,1,2,3,4]
for i in range (len(a)):
    formlist.append(a[i])
for j in range(len(b)):
    formlist.append(b[j])
for k in range(len(c)):
    formlist.append(c[k])
for l in range(len(d)):
    formlist.append(d[l])
    
print(formlist)
formlist.sort()
print(formlist,len(formlist))

actlist=[]
for x in (formlist):
    if x not in actlist:
        actlist.append(x) 

print(actlist,len(actlist))   
```
