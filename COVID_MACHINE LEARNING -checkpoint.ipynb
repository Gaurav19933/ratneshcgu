```python
India_case=covid[covid["location"]=='India']
```

```python
India_case.head()
```

```python
India_case_5=India_case.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=India_case)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=India_case_5)
plt.show()
```

```python
China_case=covid[covid["location"]=='China']
Germany_case=covid[covid["location"]=='Germany']
USVI_case=covid[covid["location"]=='United States Virgin Islands']
```

```python
China_case_5=China_case.tail()
Germany_case_5=Germany_case.tail()
USVI_case_5=USVI_case.tail()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=China_case)
sns.lineplot(x="date",y="total_cases",data=India_case)
sns.lineplot(x="date",y="total_cases",data=Germany_case)
sns.lineplot(x="date",y="total_cases",data=USVI_case)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=China_case_5)
sns.lineplot(x="date",y="total_cases",data=India_case_5)
sns.lineplot(x="date",y="total_cases",data=Germany_case_5)
sns.lineplot(x="date",y="total_cases",data=USVI_case_5)
plt.show()
```
