# Linear regression and error analysis with Statistical Method

```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import seaborn as sn

#x=[1,2,3,4,5]
#y=[3,4,2,4,5]

x=np.arange(10)
#y=np.arange(0,20,2)
y=rd.sample(range(0,20),10)

# //Equation of line//  y=mx+c

#//Regression line //   ymean=m*xmean+c
# m= sum of range[(x-xmean)*(y-ymean)]/sum of range [(x-xmean)**2]
#c=y_mean-m_xmean

xmean=np.mean(x)
ymean=np.mean(y)
xiyi=[]
xi_2=[]
y_reg=[]
for i in range(len(x)):
    xiyi.append((x[i]-xmean)*(y[i]-ymean))
    xi_2.append((x[i]-xmean)**2)
xiyi_sum=np.sum(xiyi)
xi_2sum=np.sum(xi_2)
m=(xiyi_sum)/(xi_2sum) #slope
c=ymean-m*xmean

print("slope_m",m)
print("xmean:",xmean,"ymean:",ymean)
print(xmean*m)
print("constant_c",c)


# plot the regression line obtained with the values // it is called predicted line //

for i in range(len(x)):
    y_reg.append(m*x[i]+c)

# Error estimation in linear regression is calculated by R_Square method
# r**2=sum of range(y_reg-ymean)**2/sum of range(y-ymean)**2

num=[]
den=[]
for i in range (len(y)):
    num.append((y_reg[i]-ymean)**2)
    den.append((y[i]-ymean)**2)
num1=np.sum(num)
den1=np.sum(den)
r=(num1**2/den1**2)**(0.5)
print("R_error",r)
sn.set(rc={'figure.figsize':(15,10)})
sn.scatterplot(x,y)
#sn.lineplot(x,y)
sn.scatterplot(x,y_reg)
sn.lineplot(x,y_reg)
#plt.plot(x,y)
plt.show()
```
